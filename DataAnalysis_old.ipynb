```python
import os
import glob
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

```python
wifi_path = "Datasets/WiFi dataset"
mag_path = "Datasets/Magnetic field dataset"

print("WiFi folder exists:", os.path.exists(wifi_path))
print("Magnetic folder exists:", os.path.exists(mag_path))
```

```python
for root, dirs, files in os.walk("Datasets"):
    print(f"\n{root}")
    for f in files[:5]:  # show first 5 files
        print("   ", f)
```

```python
mag_files = glob.glob("Datasets/Magnetic field dataset/**/*.csv", recursive=True)

print("Total magnetic files:", len(mag_files))

sample_mag_file = mag_files[0]
print("Example file:", sample_mag_file)
```

```python
df_mag = pd.read_csv(sample_mag_file)

df_mag.head()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(df_mag["Mag_x"], bins=50, kde=True)
plt.title("Distribution of Magnetic Field X")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(df_mag["Mag_y"], bins=50, kde=True)
plt.title("Distribution of Magnetic Field X")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(df_mag["Mag_z"], bins=50, kde=True)
plt.title("Distribution of Magnetic Field X")
plt.show()
```

```python
df_mag["Mag_magnitude"] = np.sqrt(
    df_mag["Mag_x"]**2 +
    df_mag["Mag_y"]**2 +
    df_mag["Mag_z"]**2
)
```

```python
df_mag["Time"] = pd.to_datetime(df_mag["Time"])

df_mag.set_index("Time", inplace=True)

df_mag["Mag_magnitude"].plot(figsize=(12,5))
plt.title("Magnetic Field Over Time")
plt.show()
```
